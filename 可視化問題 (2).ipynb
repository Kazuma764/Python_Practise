## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

問1.png

```python
filepath = "hist_data.csv"
hist_data = pd.read_csv(filepath)

data_a = pd.Series(hist_data["aデータ"], dtype="float")
data_b = pd.Series(hist_data["bデータ"], dtype="float")

plt.title("aデータとbデータの温度に関する度数分布", fontsize=10)
plt.xlabel("温度(℃)", fontsize=10)
plt.ylabel("度数", fontsize=10)

plt.hist(data_a, bins=30, alpha=0.5, color="orange", label="aデータ")
plt.hist(data_b, bins=30, alpha=0.5, color="c", label="bデータ")
plt.legend(loc="upper right")

plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

問2.png

```python
filepath = "scatter_data.csv"
scatter_data = pd.read_csv(filepath)

class_0 = scatter_data[scatter_data["class"] == 0]
class_1 = scatter_data[scatter_data["class"] == 1]
class_2 = scatter_data[scatter_data["class"] == 2]

plt.title("クラス０ ~ ２lengthとwidthに関する散布図", fontsize=10)
plt.xlabel("length(cm)", fontsize=10)
plt.ylabel("width(cm)", fontsize=10)
plt.grid(True)

plt.scatter(
    pd.Series(class_0["length (cm)"], dtype="float"),
    pd.Series(class_0["width (cm)"], dtype="float"),
    color="r",
    label="class0"
)

plt.scatter(
    pd.Series(class_1["length (cm)"], dtype="float"),
    pd.Series(class_1["width (cm)"], dtype="float"),
    color="b",
    label="class1"
)

plt.scatter(
    pd.Series(class_2["length (cm)"], dtype="float"),
    pd.Series(class_2["width (cm)"], dtype="float"),
    color="g",
    label="class2"
)

plt.legend(loc="upper right")

plt.show()
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

問3(1).png

問3(2).png

```python
filepath = "line_data.csv"
line_data = pd.read_csv(filepath)

term = pd.Series(line_data["期間"], dtype="object")
term = term.to_list()
number = pd.Series(line_data["件数"], dtype="int")
cost = pd.Series(line_data["費用"], dtype="int")

plt.title("期間ごとの件数の推移", fontsize=10)
plt.xlabel("期間", fontsize=10)
plt.ylabel("件数", fontsize=10)
plt.grid(True)
plt.plot(
    term,
    number,
    marker="o"
)
plt.xticks(rotation=30)

plt.show()

plt.title("期間ごとの費用の推移", fontsize=10)
plt.xlabel("期間", fontsize=10)
plt.ylabel("費用(円)", fontsize=10)
plt.grid(True)

plt.bar(
    term,
    cost
)
plt.xticks(rotation=30)


plt.show()
```
